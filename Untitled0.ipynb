```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('mall-customers-data.csv')

# Select features for clustering
features = ['annual_income', 'spending_score']
X = df[features].values

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
# These parameters can be tuned based on the desired clustering outcome
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

# Add cluster labels to the dataframe
df['Cluster'] = clusters

# Create visualization
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['annual_income'], df['spending_score'],
                     c=df['Cluster'], cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering of Mall Customers')
plt.colorbar(scatter, label='Cluster')

# Print summary statistics
print("\nNumber of clusters found:", len(set(clusters)) - (1 if -1 in clusters else 0))
print("Number of noise points:", list(clusters).count(-1))

# Print cluster statistics
print("\nCluster Statistics:")
for cluster in sorted(set(clusters)):
    cluster_size = len(df[df['Cluster'] == cluster])
    avg_income = df[df['Cluster'] == cluster]['annual_income'].mean()
    avg_spending = df[df['Cluster'] == cluster]['spending_score'].mean()

    if cluster == -1:
        print(f"\nNoise points:")
    else:
        print(f"\nCluster {cluster}:")
    print(f"Size: {cluster_size}")
    print(f"Average Income: ${avg_income:.2f}k")
    print(f"Average Spending Score: {avg_spending:.2f}")

plt.show()
```
